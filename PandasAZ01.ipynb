```python
import numpy as np 
import random 
# Series manipulation 
bool_list = []
for i in range(891):
    a = random.choice([True,False]) 
    bool_list.append(a)


pandas_boolean_series = pd.Series(bool_list,name='Age')
print(pandas_boolean_series)


#Lets access values from the dataframe using the custom series 
titanic_custom_age = titanic_train_df[pandas_boolean_series]
print(titanic_custom_age)
```
