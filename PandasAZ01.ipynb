Creating a **Dataframe**

```python
import pandas as pd 
```

```python
df = pd.DataFrame(
    {
        'Name':['Mubin UIC', 'Mehadi Hasan', 'Dogibullah N'],
        'Age':[15, 18, 113],
        'Sex':['male', 'male', 'male']
    }
)
df
```

Creating a **Series** <br>
Every column in a data frame is a <i>Series<i>

```python
#Accessing a column 
df['Name']
```

```python
# We can create a series from the scratch in pandas 
ages = pd.Series([22, 24, 25], name= 'Age', index=['UIC', 'Mubin', 'Stark'])
ages_i = pd.Series([22, 24, 25], name= 'Age') 
print(ages)
print(ages_i) # without defining the index
```

DataFrame **functions** min max etc 

```python
# Finding out the max value from pandas series 
max_age = df['Age'].max()
min_age = df['Age'].min()
print(max_age)
print(min_age)
```

**Describe**  describe() method is a powerful  method in pandas<br>
It gives a quick overview of **numerical overview** of the data frame 


1.   Number of items (Numerical)
2.   Mean
3.   Standard Deviation 
4.   min , max 
5.   percentiles 



```python
df.describe()
```
