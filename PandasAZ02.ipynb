```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
air_quality_df = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_no2.csv', index_col=0, parse_dates=True)
air_quality_df.head()
```

We can plot dataframes directly 

```python
# After index = 0 and parsing dates 
air_quality_df.plot()
```

```python
# We can plot specific column 
air_quality_df['station_paris'].plot()
```

```python
# Scatter plot : Comparing NO2 values 
air_quality_df.plot.scatter(x='station_london', y = 'station_paris', alpha=0.5,color='orange')
```

**Available methods** for plotting 

```python
[
    method_name
    for method_name in dir(air_quality_df.plot)
    if not method_name.startswith("_")
]
```

```python
air_quality_df.plot.box()
```

```python
#Area plotting with sub plots 
axs = air_quality_df.plot.area(figsize=(12, 10), subplots=True)
```

```python
fig, axs = plt.subplots(figsize=(12, 8))
air_quality_df.plot.area(ax=axs)
axs.set_ylabel("NO$_2$ concentration")
fig.savefig('my_concentrations.png')
```
