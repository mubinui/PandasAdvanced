<h2>Statistics</h2> <br> Statistics in <b>pandas 

```python
#Aggregating statistics 
titanic_train_df[['Age','Fare']].mean()
```

```python
#Lets find out the median of age and ticket price from the dataframe 
titanic_train_df[['Age', 'Fare']].median()
```

The **Describe** method for the short description of all the data 

```python
titanic_train_df[['Age','Fare']].describe()
```

We can get specific <font color='Red'> <b> combinations of aggregating statistics </b> </font> for given column 

```python
# We can do the same thing with specific methods
titanic_train_df.agg(
    {
    "Age":['min', 'max', 'median', 'skew'],
    "Fare":['min', 'max', 'median', 'skew']
    }
)
```

<h1>Aggregating statistics with <font color='Red'><b>grouped by</b></font> category</h1>

```python
titanic_train_df[["Sex", "Age"]].groupby("Sex").mean()
```

```python
titanic_train_df[['Name', 'Sex', 'Age','Pclass']].groupby('Pclass').mean()
```
